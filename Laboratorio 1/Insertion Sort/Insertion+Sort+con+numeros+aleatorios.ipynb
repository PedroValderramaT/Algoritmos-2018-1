#Insertion Sort con numeros aleatorios

llamamos las librerias

```python
import random
import matplotlib.pyplot as plt
import itertools
```

definimos la funcion de Insertion Sort y la de promedios

```python
def insertion_sort(lista):
    """toma una lista como entrada, y toma cada elemento de la lista y lo compara con una llave; retorna
    los intercambios, instrucciones y swaps"""
    
    swaps = 0
    comparisons = 0
    instructions = 0
    
    for j in range(len(lista)-1):
        llave = lista[j+1]
        i = j
        instructions += 3
        swaps += 2
        while i >= 0 and lista[i] > llave:
            lista[i+1] = lista[i]
            i -= 1
            swaps += 1
            comparisons += 2
            instructions += 3
        lista[i+1] = llave
        swaps += 1
        instructions += 1
    return swaps, comparisons, instructions

def average(lista):
    result = 0
    for i in range(len(lista)):
        result += lista[i]
    return result / len(lista)   

```

generamos datos para sacar promedio y distribucion

```python
number_of_avg_instr = [ ]
number_of_avg_comp = [ ]
number_of_avg_swaps = [ ]
avg_instr = []
avg_comp = []
avg_swaps = []

for i in range( 2, 11 ):
    k = i 
    swaps = [ ]
    comparaciones = [ ]
    instrucciones = [ ]
    numero_random = random.randint(1, i)
    m = list(itertools.permutations(range(k), k))

    for x in range (1, numero_random + 1) :
        segundo_numero_random = random.randint(0, len(m)-1)
        tuples = list(list(m[segundo_numero_random]))
        result = insertion_sort(tuples)
        swaps.append(result[0])
        comparaciones.append(result[1])
        instrucciones.append(result[2])

    number_of_avg_instr.append(instrucciones)
    number_of_avg_comp.append(comparaciones)
    number_of_avg_swaps.append(swaps)
    avg_instr.append(average(instrucciones))
    avg_comp.append(average(comparaciones))
    avg_swaps.append(average(swaps))
```

graficamos

```python
lista_instr = []
for i in range(len(number_of_avg_instr)):
    lista_instr += number_of_avg_instr[i]
    
lista_comp = []
for i in range(len(number_of_avg_comp)):
    lista_comp += number_of_avg_comp[i]

lista_swaps = []
for i in range(len(number_of_avg_swaps)):
    lista_swaps += number_of_avg_swaps[i]

    
instrucciones = []
cantidad_instrucciones = []
a = min(lista_instr)
b = max(lista_instr)
for i in range(a, b+1):
    instrucciones.append(i)
    cantidad_instrucciones.append(lista_instr.count(i))
    i += 1
    
comparaciones = []
cantidad_comparaciones = []
a = min(lista_comp)
b = max(lista_comp)
for i in range(a, b+1):
    comparaciones.append(i)
    cantidad_comparaciones.append(lista_comp.count(i))
    i += 1

swaps = []
cantidad_swaps = []
a = min(lista_swaps)
b = max(lista_swaps)
for i in range(a, b+1):
    swaps.append(i)
    cantidad_swaps.append(lista_swaps.count(i))
    i += 1
```

##Promedios

```python
plt.plot(avg_instr)
plt.plot(avg_comp)
plt.plot(avg_swaps)
plt.legend(['Instrucciones Promedio', 'Comparaciones Promedio', 'Swaps Promedio']);
```

##Distribuciones

```python
plt.plot(instrucciones,cantidad_instrucciones)
plt.xlabel('Numero instrucciones');
plt.ylabel('Repeticion de instrucciones');
```

```python
plt.plot(comparaciones,cantidad_comparaciones)
plt.xlabel('Numero comparaciones');
plt.ylabel('Repeticion de comparaciones');
```

```python
plt.plot(swaps,cantidad_swaps)
plt.xlabel('Numero swaps');
plt.ylabel('Repeticion de swaps');
```
