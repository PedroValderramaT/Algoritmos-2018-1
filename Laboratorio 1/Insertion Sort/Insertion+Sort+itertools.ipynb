# Insertion Sort con itertools

Aca llamamos a la libreria y creamos el metodo para el algoritmo de Insertion Sort

```python
import itertools

def insertion_sort(lista):
    """toma una lista como entrada, y toma cada elemento de la lista y lo compara con una llave; no retorna nada"""

    for j in range(len(lista)-1):
        llave = lista[j+1]
        i = j
        while i >= 0 and lista[i] > llave:
            lista[i+1] = lista[i]
            i -= 1
        lista[i+1] = llave
    return None
```

Aca se ve los resultados para un n = 11, despues de ese valor la memoria se llena y no sirve el codigo

```python
n = 1

while n < 11:
    lista_ord = []
    lista_a_ord = list(itertools.permutations([x for x in range(1,n+1)], n))
    for i in lista_a_ord:
        i = list(i)
        insertion_sort(i)
        lista_ord.append(i)
    n+=1
    print lista_ord  
print lista_ord   
print n
    
```
