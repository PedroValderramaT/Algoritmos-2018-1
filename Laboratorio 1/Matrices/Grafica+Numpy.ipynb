```python
import time
import numpy as np
import matplotlib.pyplot as plt

tiempos = []
n = [100,200,400,500]
for i in range(len(n)):
    suma = 0
    for j in range(10): 
        matrizA = np.ones((n[i],n[i]))
        matrizB = np.full((n[i],n[i]),2)
        A = np.mat(matrizA)
        B = np.mat(matrizB)
        tiempo_inicial = time.time()
        C = A*B
        tiempo_final = time.time()
        suma = suma + tiempo_final
        suma = suma/10
    tiempos.append(suma)
```

```python
plt.plot(n,tiempos)
plt.xlabel("n")
plt.ylabel("tiempos")
plt.show()
```
