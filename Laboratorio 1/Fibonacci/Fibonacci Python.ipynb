# Codigo Fibonacci

Primero importamos las librerias

```python
import gc
import time
import matplotlib.pyplot as plt
import numpy as np
```

funcion iterativa de python

```python
def fibo(n):
  if n == 0:
    return 0
  if n == 1 or n == 2:
    return 1
  first = 1
  second = 1
  for i in range(2,n+1):
    first , second = second + first, first
  return second
```

se crea una lista con los tiempos para cada n de fibo(n)

```python
lista_de_tiempos = []
for i in range(1000):
    gc.collect()
    tiempo_inicial = time.time()
    fibo(i)
    tiempo_final = time.time()
    gc.collect()
    lista_de_tiempos.append(tiempo_final - tiempo_inicial)
```

se hace la grafica de n = 0..1000 para 1000 tiempos

```python
n = np.arange(0, 1000)
plt.plot(n,lista_de_tiempos)
plt.xlabel("n")
plt.ylabel("tiempo")
plt.show()
```
