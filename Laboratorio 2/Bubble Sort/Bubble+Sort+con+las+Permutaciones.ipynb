#Bubble Sort con permutaciones

llamamos la libreria para graficar

```python
import matplotlib.pyplot as plt
```

Aca definimos Bubble Sort con sus respectivos swaps, comparaciones e instrucciones; la funcion promedio y la funcion de intercambios

```python
def bubble_sort(b):
    swaps = 0
    comparisons = 0
    instructions = 0

    for i in range(len(b)):
           for j in range(len(b)-i-1):
                if b[j] > b[j+1]:
                    b[j], b[j+1] = b[j+1], b[j]
                    swaps += 1 
                    comparisons += 1
                    instructions += 4
                instructions += 2
    instructions += 1        
    return swaps, comparisons, instructions

def average(lista):
    result = 0
    for i in range(len(lista)):
        result += lista[i]
    return result / len(lista)   

def swap(lista, i, j):
    lista[i], lista[j] = lista[j], lista[i]
    
   
```

Aca tomamos el codigo de las permutaciones

```python
def permutations(lista, i, j):
    cases = 0
    k = 0
    results = []
    if i == j:
        total = [ ]
        cases += 1
        for k in range( 0, j ):
            total.append(lista[k])
        results += bubble_sort(total)
        average_instructions.append(results[0])
        average_comparisons.append(results[1])
        average_swaps.append(results[2])

    else:
        for k in range(i, j):
            swap(lista, i, k)
            permutations(lista, i+1, j)
            swap(lista, i, k)
```

Aca generamos los datos para sacar el promedio y las distribuciones

```python
number_of_avg_instr = [ ]
number_of_avg_comp = [ ]
number_of_avg_swaps = [ ]
avg_instr = []
avg_comp = []
avg_swaps = []
for k in range (1,12):
    average_instructions = [ ]
    average_comparisons = [ ]
    average_swaps = [ ]

    list_of_num = range(k)
    list_of_permutations = permutations(list_of_num, 0, k )
    
    number_of_avg_instr.append(average_instructions)
    number_of_avg_comp.append(average_comparisons)
    number_of_avg_swaps.append(average_swaps)
    avg_instr.append(average(average_instructions))
    avg_comp.append(average(average_comparisons))
    avg_swaps.append(average(average_swaps))
```

Graficamos

```python
lista_instr = []
for i in range(len(number_of_avg_instr)):
    lista_instr += number_of_avg_instr[i]
    
lista_comp = []
for i in range(len(number_of_avg_comp)):
    lista_comp += number_of_avg_comp[i]

lista_swaps = []
for i in range(len(number_of_avg_swaps)):
    lista_swaps += number_of_avg_swaps[i]

    
instrucciones = []
cantidad_instrucciones = []
a = min(lista_instr)
b = max(lista_instr)
for i in range(a, b+1):
    instrucciones.append(i)
    cantidad_instrucciones.append(lista_instr.count(i))
    i += 1
    
comparaciones = []
cantidad_comparaciones = []
a = min(lista_comp)
b = max(lista_comp)
for i in range(a, b+1):
    comparaciones.append(i)
    cantidad_comparaciones.append(lista_comp.count(i))
    i += 1

swaps = []
cantidad_swaps = []
a = min(lista_swaps)
b = max(lista_swaps)
for i in range(a, b+1):
    swaps.append(i)
    cantidad_swaps.append(lista_swaps.count(i))
    i += 1
    
```

##Promedios

```python
plt.plot(avg_instr)
plt.plot(avg_comp)
plt.plot(avg_swaps)
plt.legend(['Instrucciones Promedio', 'Comparaciones Promedio', 'Swaps Promedio']);
```
